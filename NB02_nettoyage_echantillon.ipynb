# Exploration des valeurs manquantes

```python
# calcul du taux de valeurs manquantes dans le data set

datas.isnull().sum().sum()/datas.size
```

```python
# calcul du taux de valeurs manquantes dans les infos nutritionnelles

scored_datas.isnull().sum().sum()/scored_datas.size
```

```python
# suppression des colonnes avec moins de 500 valeurs dans nutr_features

nutr_features = [col for col in nutr_features if scored_datas[col].notnull().sum()>=500]
len(nutr_features)
```

```python
# calcul et représentation du nombre de valeurs manquantes pour chacune de ces colonnes

nb_total = scored_datas.shape[0]

for col in nutr_features :
    nb_nan = scored_datas[col].isnull().sum()
    fig = plt.figure(figsize=(5,5))
    plt.pie([nb_nan, nb_total-nb_nan], labels=["Valeurs manquantes", "Valeurs présentes"], labeldistance=None)
    plt.title(col)
    plt.legend(loc="upper right")
    plt.show()
```

```python
# figure globale

fig, axes = plt.subplots(3, 7,figsize=(22,12))
axes = axes.ravel()
for i, col in enumerate(nutr_features):
    ax = axes[i]
    nb_nan = scored_datas[col].isnull().sum()
    ax.pie([nb_nan, nb_total-nb_nan], labels=["Valeurs manquantes", "Valeurs présentes"],labeldistance=None)
    ax.set_title(col, fontsize=17)
    
fig.tight_layout()

lines, labels = fig.axes[-1].get_legend_handles_labels()
fig.legend(lines, labels, bbox_to_anchor=(0.6, 0.1), fontsize=25)

plt.savefig("valeurs_manquantes", bbox_inches="tight")
plt.show()
```

```python
# on garde les attributs ayant suffisamment de valeurs

relevant_features = ["energy-kcal_100g", 
                     "energy_100g", 
                     "fat_100g", 
                     "saturated-fat_100g", 
                     "carbohydrates_100g",
                     "sugars_100g",
                     "fiber_100g",
                     "proteins_100g",
                     "salt_100g",
                     "sodium_100g",
                     "nutrition-score-fr_100g"]

```

```python
scored_datas[relevant_features].describe()

```

On constate des valeurs aberrantes pour certains attributs (le maximum devrait être 100).

# Nettoyage et imputation des données

```python
# analyse univariée de nutrition-score-fr

target = scored_datas["nutrition-score-fr_100g"]

print(target.describe())

fig, ax = plt.subplots(1,1)
plt.boxplot(target, showmeans=True)
plt.title("Diagramme en boîte\nnutrition-score-fr_100g", fontsize=20)
ax.set_xticklabels([""])
plt.savefig("boite_cible.png", bbox_inches="tight")
plt.show()


sns.distplot(a = target, kde=True)
plt.title("Distribution", fontsize=20)
plt.savefig("distrib_cible")
```

Il n'y a pas d'outliers, la distribution est bimodale.

```python
# suppression des observations avec des valeurs clairement aberrantes (>100 dans les composants)

components = relevant_features[2:-1]

def non_aberrant(line) :
    return (max(line[components].values) < 100)

scored_datas = scored_datas[scored_datas.apply(non_aberrant, axis="columns")][relevant_features]

scored_datas
```

```python
# diagrammes en boîte des variables

fig, axes = plt.subplots(2, 5,figsize=(22,8))
axes = axes.ravel()

for i, col in enumerate([c for c in scored_datas.columns if c!="nutrition-score-fr_100g"]):
    ax = axes[i]
    ax.boxplot(scored_datas[col].dropna(), showmeans=True)
    ax.set_xticklabels([""])
    ax.set_title(col, fontsize=20)
    
fig.tight_layout()


plt.savefig("boites_variables.png", bbox_inches="tight")
plt.show()
```

On constate des outliers, mais ils sont cohérents avec l'esprit de Open Fact Food, on y trouve en effetdes produits très variés y compris du sel ou du sucre pur.  
Je conserve donc l'ensemble des données.

```python
# calcul du taux de valeurs manquantes

print(f"Taux de valeurs manquantes : {scored_datas.isnull().sum().sum()/scored_datas.size}")

print("valeurs manquantes par colonnes :")
scored_datas.isnull().sum()
```

```python
# visualisation de la matrice de corrélation

corr = scored_datas.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
)
```

On constate une corrélation importante des attributs energy-kcal_100g et energy_100g, ainsi que salt_100g et sodium_100g, ce qui est normal puisqu'ils représentent les même informations.  
Je supprime donc energy-kcal_100g qui contient des valeurs manquantes.  
Pour les valeurs manquantes de fiber_100g, je vais les imputer par les plus proches voisins puisqu'il n'y a pas de corrélation franche avec d'autres attributs.

```python
# suppression de l'attribut enegy-kcal_100g

scored_datas.drop(columns="energy-kcal_100g", inplace=True)
relevant_features.remove("energy-kcal_100g")
```

```python
# imputation par kNN

from sklearn import impute

scored_imputed_datas = scored_datas.copy()
scored_imputed_datas[relevant_features] = impute.KNNImputer().fit_transform(scored_datas)
```

```python
scored_imputed_datas.describe()
```

```python
# enregistrement des données et de la liste des attributs pertinents

scored_imputed_datas.to_csv("imputed_sample.csv", index=False)

import pickle

pickle.dump(relevant_features, open("pickle_relevant_features.pkl", "wb"))
```
